```python
import torch

log_dir="/data/targetdiff_data/PCBA_sample_output/PDBBind-DUD_E_FLAPP_0.9/eval_results/metrics_-1.pt"
metrics = torch.load(log_dir)

ligand_vina={}
for sample in metrics['all_results']:
    print(sample['ligand_filename'], sample['vina'])
    if sample['ligand_filename'] not in ligand_vina:
        ligand_vina[sample['ligand_filename']]=[]
    ligand_vina[sample['ligand_filename']].append(sample['vina']['score_only'][0]['affinity'])

# print average vina score for each ligand
for ligand in ligand_vina:
    print(ligand, sum(ligand_vina[ligand])/len(ligand_vina[ligand]))
```
