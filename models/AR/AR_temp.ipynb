```python
output_dir="/data/AR_data/sample_output/PDBBind-DUD_E_FLAPP_0.6"
import glob
import os
finish_cnt=0
item_list=glob.glob(os.path.join(output_dir, "*"))
item_list = [x for x in item_list if os.path.isdir(x)]
print(item_list)
for item in item_list:
    sdf_dir=os.path.join(item, "SDF")
    cnt=len(glob.glob(os.path.join(sdf_dir, "*.sdf")))
    if cnt>0:
        finish_cnt+=1
    print(os.path.basename(item), cnt)

print(finish_cnt)
```

```python
import glob
samples=glob.glob("/data/AR_data/sample_output/PDBBind-DUD_E_FLAPP_0.6/23_src/SDF/*.sdf")


ok_list=[]

import rdkit
from rdkit import Chem
for sample in samples:
    # check if the sdf file is valid
    try:
        mol = Chem.MolFromMolFile(sample,sanitize=False)

        # try sanitize
        Chem.SanitizeMol(mol)

        # try kekulize
        Chem.Kekulize(mol)
        ok_list.append(sample)
    except:
        print("error")
        continue
    
```
